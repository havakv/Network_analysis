## Subgraphs

$H = (V_H, E_H)$ 
is a subgraph of $G = (V_G, E_G)$ if $V_H \subseteq V_G$ and $E_H \subseteq E_G$.

An *induced subgraph* is a subgraph $G' = (V', E')$ with prespecified vertices $V' \subseteq V_G$, 
and the edges $E' \subseteq E_G$ are the collection of edges to be found om $G$ among the subset of vertices.

```r
h <- induced.subgraph(g, 1:5)
par(mfrow=c(1,2))
par(mar = c(1,1,1,1))
plot(g, vertex.size=50)
plot(h, vertex.size=50)
```

Can also use subtraction and addition operators

```r
h2 = g - vertices(c(6, 7))
par(mar = c(1,1,1,1))
plot(h2, vertex.size=50)
```

# Some basic graph concepts

Graphs without loops (edge from node to itself) and mutiple edges between vertices, are called simple. Many models and methods assume the input graph is simple. Multigraphs are often converted to weighted simple graphs instead.

```r
is.simple(g)
```

```r
par(mar = c(1,1,1,1))
plot(g, vertex.size=50)
```

### Connectivity

Two verticies are said to be *adjacent* or *neighbors* if they are joined by an edge.

```r
neighbors(g, 5)
```

The *degree* of a vertex $v$, say $d_v$, is the number of edges incident on to $v$ (how many edges connected to it). 

```r
print(degree(g))
```

For directed graphs, degree is replaces by *in-degree* and *out-degree*. 

```r
par(mar = c(1, 1, 1, 1))
plot(dg)
```

```r
print(degree(dg, mode = "in"))
```

A *cycle* is a path from one vertex to itself without repeated nodes.

A *connected graph* is a grap where every vertex is reachable from every other.

A *component* of a graph, is a maximally connected subgraph, i.e. the larges connected subgraph.

```r
is.connected(g)
```

A directed graph is *weakly* connected if its underlying (undirected) graph is connected.

It is *strongly* connected if all vertices are reachable from all vertices.

```r
is.connected(dg, mode="weak")
```

```r
is.connected(dg, mode="strong")
```

### Diameter

*Geodesic distance* is the shortest path between two nodes. 

The longest geodesic distance in a graph is defined as its *diameter*

```r
diameter(g, weights=NA)
```

## Special types of graphs

Some commonly encountered types of graphs are:

- *Complete*: All nodes are connected by an edge.
    - A *clique* is a complete subraph.
- *Regular*: Every vertex has the same degree.
- *Tree*: Connected graph with no cycles.
- *k-star*: A tree with only one root and $k$ leaves.
- *DAG*: A directed acyclic graph is a directed graph without cycles.
- *Bipartite*: A graph $G = (V, E)$, s.t. $V$ can be partitioned into two disjoint sets, $V_1$ and $V_2$. 
    - Induced graphs where the common nodes represent edges are called *projections*.

```r
g.full <- graph.full(7)
g.ring <- graph.ring(7)
g.tree <- graph.tree(7, children=2, mode="undirected")
g.star <- graph.star(7, mode="undirected")
par(mfrow=c(2, 2))
par(mar = c(1, 1, 1, 1))
plot(g.full, vertex.size=40)
title("Complete / 7-degree")
plot(g.ring, vertex.size=40)
title("Ring / 2-regular")
plot(g.tree, vertex.size=40)
title("Tree")
plot(g.star, vertex.size=40)
title("Star")
options(repr.plot.width=5, repr.plot.height=4)
```

### A bipartite graph

```r
g.bip <- graph.formula(actor1:actor2:actor3,
                       movie1:movie2, actor1:actor2 - movie1,
                       actor2:actor3 - movie2)
V(g.bip)$type <- grepl("^movie", V(g.bip)$name)
str(g.bip, v=T)
```

```r
par(mar = rep(1, 4))
plot(g.bip, layout=-layout.bipartite(g.bip)[,2:1],
     vertex.size=60, 
     vertex.shape=ifelse(V(g.bip)$type, "square", "circle"),
     vertex.color=ifelse(V(g.bip)$type, "pink", "lightblue"))
```

### Get projections

```r
proj <- bipartite.projection(g.bip)
par(mfrow = c(1, 2), mar = rep(1, 4))
plot(proj[[1]], vertex.size=100, vertex.shape="circle", vertex.color="lightblue")
plot(proj[[2]], vertex.size=100, vertex.shape="square", vertex.color="pink")
```
